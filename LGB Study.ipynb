```r
data=read.csv("analysis2.csv")
```

```r
data <- data %>%
  gather(key = "subscale", value = "score",rel,fc2,cc,sid,acccon,conmot,ideunc,inthom,diffpro,idesup,ideaff,idecen) %>%
  convert_as_factor(subscale)
```

```r
head(data)
```

```r
library("viridis") 

dp <- ggplot(data, aes(x=mingro, y=score, fill=mingro2)) + 
  geom_violin(trim=FALSE)+ geom_boxplot(width=0.1, fill="white")+
  labs(x="",y = "Score")
dp=dp + scale_fill_brewer(palette="heat",name="",labels = c( "No Minority","Caste-/Religion-based Minority")) + scale_x_discrete(labels=c("yes" = "C/R Minority", "no" = "No Minority"))
hospital_names <- list(
  'acccon'="Acceptance Concerns",
  'fc2'="Family Connectedness",
  'rel'="Religiosity", 
  'cc'="Community Connectedness", 
  'sid'="Sexual Identity Distress",
  'conmot'="Concealment Motivation",
  'ideunc'="Identity Uncertainty",
  'inthom'="Internalized Homonegativity",
  'diffpro'="Difficult Process",
  'idesup'="Identity Superiority",
  'ideaff'="Identity Affirmation",
  'idecen'="Identity Centrality")


hospital_labeller <- function(variable,value){
  return(hospital_names[value])
}

dp + facet_wrap(subscale ~ ., ncol=4, labeller=hospital_labeller)+ theme(legend.position="bottom")
```
